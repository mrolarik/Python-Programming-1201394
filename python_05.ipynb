## Pandas สร้าง DataFrame

````
import pandas as pd

std_info = pd.DataFrame(
    {
        'id' : ['001', '002', '003','004'],
        'Name' : ['Olarik', 'Kaveepoj', 'Suwich', 'Suwichai'],
        'Dep' : ['Information Technology', 'Computer Science', 'New Media', 'New Media'],
        'Height' : [165, 185, 165, 160],
        'Weight' : [75, 105, 60, 55],
    }
)

# แสดงข้อมูลใน DataFrame
std_info
````

### แสดงข้อมูลใน DataFrame

````
# แสดงตั้งแต่ลำดับที่ 1 ถึงลำดับสุดท้าย
std_info[1:]
````

````
# แสดงตั้งแต่ลำดับแรกถึงลำดับที่ 1
std_info[:2]
````

### การจัดเรียงข้อมูลใน DataFrame

````
# เรียงข้อมูลตาม 'Name'
std_info.sort_values(by=['Name'])
````

````
# เรียงข้อมูลตาม 'Name' เรียงจากมากไปหาน้อย
std_info.sort_values(by=['Name'], ascending=False)
````

````
# เรียงข้อมูลตาม 'Dep' และ 'Height' และกำหนดให้เรียงจากน้อยไปมาก
std_info.sort_values(by=['Dep', 'Height'], ascending=True)
````

## ค้นหาข้อมูลใน DataFrame - Query

ค้นหาข้อมูลประเทศ (Country) 'United Kingdom' และแสดงจำนวนข้อมูลที่ค้นพบ

````
result = df[df['Country'] == 'United Kingdom']

print('Found: ', result.shape[0], 'rows')
result
````

ค้นหาข้อมูลประเทศ (Country) 'United Kingdom' ที่จ่ายเงินด้วยบัตร (Payment_Type) ประเภท 'Visa' และแสดงจำนวนข้อมูลที่ค้นพบ

````
result = df[df['Country'] == 'United Kingdom']
result = result[df['Payment_Type'] == 'Visa']

print('Found: ', result.shape[0], 'rows')
result
````

````
result = df[df['Country'] == 'United Kingdom'][df['Payment_Type'] == 'Visa']

print('Found: ', result.shape[0], 'rows')
result
````

````
result = df[ (df['Country'] == 'United Kingdom') & (df['Payment_Type'] == 'Visa')  ]

print('Found: ', result.shape[0], 'rows')
result
````

ค้นหาข้อมูลประเทศ (Country) 'United Kingdom' ที่จ่ายเงินด้วยบัตร (Payment_Type) ประเภท 'Visa' และจ่ายเงิน (Price) มากกว่า 1200 บาท และแสดงจำนวนข้อมูลที่ค้นพบ

````
result = df[ (df['Country'] == 'United Kingdom') & (df['Payment_Type'] == 'Visa') & (df['Price'] > str(1200)) ]

print('Found: ', result.shape[0], 'rows')
result
````

หาความแตกต่างของการใช้จ่ายผ่านบัตร Visa และ Mastercard

````
import pandas as pd

df = pd.read_csv('http://www.sharecsv.com/dl/9096d32f98aa0ac671a1cca16fa43be8/SalesJan2009.csv')
df['Price'] = df['Price'].replace('[,]','', regex=True).astype(int)

# visa
visa = df[df['Payment_Type'] == 'Visa']
visa = visa['Price'].sum()

mastercard = df[df['Payment_Type'] == 'Mastercard']
mastercard = mastercard['Price'].sum()

print('Payment Type')
print('Visa:', visa, '$')
print('mastercard:',mastercard, '$')
print('Different:', visa-mastercard, '$')
````

# การค้นหาข้อมูลด้วย Pandas - Search

````
import pandas as pd

df = pd.read_csv('http://www.sharecsv.com/dl/9096d32f98aa0ac671a1cca16fa43be8/SalesJan2009.csv')
````

````
df.head()
````

## ใช้ฟังก์ชัน startswith เพื่อค้นหาข้อมูล

### ค้นหา 'City' ที่ขึ้นต้นด้วย 'Bas'

````
df['City'].str.startswith('Bas')
````

````
df[df['City'].str.startswith('Bas')]
````

### ปัญหาของ case sensitive ที่พบ

````
# กรณีนี้ค้นหาไม่เจอ เนื่องจากไม่มีเมืองไหนที่ขึ้นต้นด้วย 'as' มีแต่ 'AS'
# case sensitive
df[df['City'].str.startswith('as')]
````

````
df[df['City'].str.startswith('As')]
````

## ใช้ฟังก์ชัน contains ในการค้นหา

````
# ค้นหาข้อมูลที่ประกอบด้วยตัวอักษร (contains) ที่ค้นหา 
# เช่น ค้นหาคำว่า 'as' ดังนั้น จึงค้นพบ 'Basildon', 'Gold Coast' 
df[df['City'].str.contains('as', regex=True)]
````

````
# กรณีการค้นหาจะเป็น case sensitive ดังนั้น 'as'กับ 'AS' มีความหมายที่แตกต่างกัน
df[df['City'].str.contains('As', regex=True)]
````

#### ป้องกันปัญหา case sensitive ในฟังก์ชัน contains


````
# จากตัวอย่างค้นเจอทั้ง 'Basildon' และ 'Astoria' ที่มี 'as' และ'As' อยู่ในคำ
# (?i) เป็น regex pattern ที่ใช้จัดการกับ case sensitive (ignore case sensitive)

result = df[df['City'].str.contains('(?i)as', regex=True)]
print(result.shape)
result
````

````
result = df[df['City'].str.contains('as', case=False)]
print(result.shape)
result

# หรือ
#result = df[df['City'].str.contains('as|recess', case=False)]
#print(result.shape)
````
